```python
def plot_pq_parameters(file_path, save_path=None):
    """
    读取文件并绘制 p 和 q 参数的变化曲线图。

    :param file_path: 包含 Best Parameters 的 CSV 文件路径
    :param save_path: (可选) 保存图像的路径，如果不指定则不保存
    :param dpi: 图像的分辨率，默认值为300
    """
    # 读取文件
    Best_params = pd.read_csv(file_path)

    # 提取第一个参数 p 和第二个参数 q
    p_values = Best_params['Best Parameters'].apply(lambda x: float(x.strip('()').split(',')[0]))
    q_values = Best_params['Best Parameters'].apply(lambda x: float(x.strip('()').split(',')[1]))

    # 创建图和轴
    fig, ax1 = plt.subplots(figsize=(8, 6))

    # 绘制第一个参数 p 在左侧 y 轴
    ax1.plot(Best_params['Time Step'] + 1, p_values, marker='o', linestyle='-', color='#ff0000', label='p')
    ax1.set_xlabel('时间区间', fontsize=16, fontproperties=font_prop)
    ax1.set_ylabel('p', color='#ff0000', fontsize=16, fontname='Times New Roman')
    ax1.tick_params(axis='y', labelcolor='#ff0000')
    ax1.tick_params(axis='both', which='major', labelsize=16)
    for label in ax1.get_xticklabels() + ax1.get_yticklabels():
        label.set_fontname('Times New Roman')

    # 创建第二个 y 轴用于绘制 q 参数
    ax2 = ax1.twinx()
    ax2.plot(Best_params['Time Step'] + 1, q_values, marker='s', linestyle='--', color='#4169e0', label='q')
    ax2.set_ylabel('q', color='#4169e0', fontsize=16, fontname='Times New Roman')
    ax2.tick_params(axis='y', labelcolor='#4169e0')
    ax2.tick_params(axis='both', which='major', labelsize=16)
    for label in ax2.get_xticklabels() + ax2.get_yticklabels():
        label.set_fontname('Times New Roman')

    # 调整布局以便图例显示
    plt.tight_layout()

    # 添加图例，字体为 Times New Roman
 
    fig.legend(loc='lower right', bbox_to_anchor=(0.75, 0.15), fontsize=16)

    # 保存图像（如果指定了保存路径）
    if save_path:
        plt.savefig(save_path)

    # 显示图像
    plt.show()

# 示例用法：
# plot_pq_parameters('D:/研究生/毕业论文/代码/网络构建/直播平台网络模型/python/Multi_dynamic_community_network Project/实验结果/params_and_distances_r=50_l1.csv',
#                    save_path='D:/lunwen/social_network2023.7.17/caogao/paper0_2024.8.5/figures/pq_50.pdf')
```

```python
# 创建用于存储解析结果的列表
time_steps = []

with open(log_file, 'r') as file:
    lines = file.readlines()
    for line in lines:
        # 使用正则表达式提取时间步、参数和曼哈顿距离信息
        match = re.search(r'Time Step (\d+): Parameters: \(([^)]+)\), MAE: ([\d.]+)', line)
        if match:
            time_step = int(match.group(1))
            
            # 将解析的数据添加到列表中
            time_steps.append(time_step)
```

```python
from sklearn.model_selection import train_test_split
LIVE_DATA_PATH = 'D:/lunwen/social network2023.7.17/daima/zhibo/新实验/Jupyter code/live_data_distribution.csv'
CHANGE_DATA_PATH = 'D:/lunwen/social network2023.7.17/daima/zhibo/新实验/Jupyter code/Nt_change.csv'

DESIRED_COLUMNS = ['real_time_user', 'leave_user', 'user_count']
DESIRED_COLUMNS1 = [f'{col}_t' for col in DESIRED_COLUMNS]
MAX_TIME_POINT = 60
PARAM_VALUES_P = np.arange(0, 1.1, step=0.1).round(1)
PARAM_VALUES_Q = np.arange(-3, 3.1, step=0.2).round(1)
RANDOM_SEED = 43

# Load data
live_data_distribution = pd.read_csv(LIVE_DATA_PATH, index_col=0)
live_data_distribution = live_data_distribution.astype(int)
Nt_change = pd.read_csv(CHANGE_DATA_PATH, index_col=0)
# 所有值加1
Nt_change += 1

# Generate time points
time_points = np.arange(0, MAX_TIME_POINT)

# Prepare train and test datasets
train_datasets = []
test_datasets = []

for t in time_points:
    data = live_data_distribution[live_data_distribution.index == t]
    train_data, test_data = train_test_split(data, test_size=0.3, random_state=RANDOM_SEED)
    train_datasets.append(train_data)
    test_datasets.append(test_data)


# 最大最小规范化函数
def min_max_normalize(series):
    min_val = series.min()
    max_val = series.max()
    if min_val == max_val:
        return pd.Series([0.0] * len(series), index=series.index).round(6)
    normalized_series = (series - min_val) / (max_val - min_val)
    return normalized_series.round(6)



# 对每个训练集和测试集进行最大最小规范化
normalized_train_datasets = []
normalized_test_datasets = []
normalized_nt_change = []

for train_data, test_data in zip(train_datasets, test_datasets):
    normalized_train_data = train_data.copy()
    normalized_test_data = test_data.copy()

    for column in train_data.columns:  # 忽略'time'列
        normalized_train_data[column] = min_max_normalize(train_data[column])
        normalized_test_data[column] = min_max_normalize(test_data[column])

    normalized_train_datasets.append(normalized_train_data)
    normalized_test_datasets.append(normalized_test_data)

# 对 Nt_change 进行规范化
normalized_nt_change = Nt_change.copy()
for column in Nt_change.columns:
    normalized_nt_change[column] = min_max_normalize(Nt_change[column])
```

```python
def simulate_audience_increment(distribution_x, change_data, t, p, q):

    increment_total_user = change_data.iloc[t, 0]
    total_change = distribution_x.iloc[0] * increment_total_user
    all_audience = total_change + distribution_x.iloc[2]

    # 初始化增量分布
    # 初始化 increment_distribution
    increment_distribution = [0.0] * len(distribution_x)
    # 精确到小数点后六位
    increment_distribution = [round(num, 6) for num in increment_distribution]
    increment_distribution[0] = total_change

    # 随机加入总观众数
    random_join_total = all_audience * p
    # 计算每个社区的加入观众数
    random_join_community1 = random_join_total * 0.5
    random_join_community2 = random_join_total - random_join_community1
    # 更新社区1的观众数
    increment_distribution[1] += random_join_community1
    # 更新社区2的观众数
    increment_distribution[2] += random_join_community2

    probabilities = np.array([distribution_x.iloc[i] ** q for i in range(1, 3)])
    # 将inf对应的概率设置为1，其他值设置为0
    probabilities = np.where(np.isinf(probabilities), 1.0, 0.0)
    # 如果probabilities都为0，则取(0.5, 0.5)
    if np.sum(probabilities) == 0:
        p_list = [0.5, 0.5]
    else:
        p_list = probabilities / np.sum(probabilities)

    # 偏好加入总观众数
    preference_join_total = all_audience * (1 - p)

    # 计算每个社区的加入观众数
    preference_join_community1 = preference_join_total * p_list[0]
    preference_join_community2 = preference_join_total - preference_join_community1
    # 更新社区1的观众数
    increment_distribution[1] += preference_join_community1
    # 更新社区2的观众数
    increment_distribution[2] += preference_join_community2

    return increment_distribution


# Function to calculate Mean Absolute Error (MAE)
def calculate_mae(vector1, vector2):
    return np.mean(np.abs(np.array(vector1) - np.array(vector2)))


# Function to evaluate and optimize parameters
def evaluate_and_optimize_parameters(p, q, t, train_data_t):
    mae_list = []

    for index, row in train_data_t.iterrows():
        true_distribution = row[DESIRED_COLUMNS1]
        predicted_distribution = simulate_audience_increment(row[DESIRED_COLUMNS], normalized_nt_change, t, p, q)

        # Calculate Mean Absolute Error
        mae = calculate_mae(predicted_distribution, true_distribution)
        mae_list.append(mae)

    # Calculate the average MAE
    average_mae = np.mean(mae_list)

    return average_mae
```
